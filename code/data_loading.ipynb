```python
import pandas as pd
```

```python
raw = pd.read_csv('http://hilpisch.com/pyalgo_eikon_eod_data.csv',
                            index_col=0, parse_dates=True).dropna()
```

```python
raw.info()
```

```python
data = pd.DataFrame(raw['EUR='])
```

```python
data.info()
```

```python
data.rename(columns={'EUR=': 'price'}, inplace = True)
```

```python
data.info()
```

```python
data['SMA1'] = data['price'].rolling(42).mean()
data['SMA2'] = data['price'].rolling(252).mean()
data.tail()
```

```python
%matplotlib inline
```

```python
from pylab import mpl, plt
import seaborn as sns
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
sns.set_theme()
```

```python
data.plot(title='EUR/USD | 42 & 252 days SMAs', figsize=(10,6))
```

```python
import numpy as np
```

```python
#strategy: go long when shorter SMA is larger than longer SMA and vice-versa

data['position'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
data.dropna(inplace = True)
data['position'].plot(ylim=[-1.1, 1.1], title='Market Positioning', figsize=(10,6))
```

```python
#RETURNS
data['returns'] = np.log(data['price'] / data['price'].shift(1))
data['returns'].hist(bins=35, figsize=(10,6))
```

```python
#comparing returns over passive benchmark strategy

data['strategy'] = data['position'].shift(1) * data['returns'] #log returns of strategy given positions & market returns
data[['returns','strategy']].sum()
```

```python
data[['returns', 'strategy']].sum().apply(np.exp) #gross performance
```

```python
#cumulative returns
data[['returns', 'strategy']].cumsum().apply(np.exp).plot(figsize=(10,6))
plt.title('gross performance of EUR/USD compared to SMA based strategy')
plt.show()
```

```python
#average annualized risk-return statistics
data[['returns', 'strategy']].mean() * 252
```

```python
np.exp(data[['returns', 'strategy']].mean() * 252) -1
```

```python
data[['returns', 'strategy']].std() * 252 ** 0.5
```

```python
(data[['returns', 'strategy']].apply(np.exp) -1 ).std() * 252 ** 0.5
```

```python
#maximum drawdown and longest drawdown
```

```python
data['cumret'] = data['strategy'].cumsum().apply(np.exp)
data['cummax'] = data['cumret'].cummax()
data[['cumret', 'cummax']].dropna().plot(figsize = (10,6))
```

```python
drawdown = data['cummax'] - data['cumret']
```

```python
drawdown.max()
```

```python
#longest drawdown period

temp = drawdown[drawdown == 0]
periods = (temp.index[1:].to_pydatetime() - temp.index[:-1].to_pydatetime())
```
